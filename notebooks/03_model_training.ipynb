```python
import pandas as pd

# 1. Cargar datos
df_y = pd.read_csv('../data/raw/target_spei.csv')
df_x = pd.read_csv('../data/raw/features_sentinel.csv')

# Formatear fechas a meses (porque el SPEI suele venir mensual)
df_y['Fecha'] = pd.to_datetime(df_y['Fecha']).dt.to_period('M')
df_x['Fecha'] = pd.to_datetime(df_x['Fecha']).dt.to_period('M')

# Agrupamos el satélite (que era semanal) a media mensual para que encaje con el SPEI
df_x_mensual = df_x.groupby(['Zona', 'Fecha']).mean().reset_index()

# 2. El Merge (Unir X e Y)
df = pd.merge(df_x_mensual, df_y, on=['Zona', 'Fecha'], how='inner')

# 3. LOS TIME-LAGS (Tu libreta en código)
df = df.sort_values(by=['Zona', 'Fecha'])

# Agrupamos por zona ANTES de hacer el shift para no mezclar ciudades
df['NDMI_hace_1_mes'] = df.groupby('Zona')['NDMI'].shift(1)
df['NDMI_hace_2_mes'] = df.groupby('Zona')['NDMI'].shift(2)

# Quitamos los nulos generados por el shift
df = df.dropna()

print(df.head()) # ¡MIRA ESTA BELLEZA DE TABLA!
```
